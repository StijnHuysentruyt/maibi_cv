# Image histogram

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_bw = cv2.imread('data/verona_bw.png', cv2.IMREAD_GRAYSCALE)

fig_bw, ax_bw = plt.subplots()
ax_bw.hist(img_bw.flatten(), bins=256, color='black');
```

```python
img_gray = cv2.imread('data/verona_gray.jpg', cv2.IMREAD_GRAYSCALE)

fig_gray, ax_gray = plt.subplots()
ax_gray.hist(img_gray.flatten(), bins=256, color='black');
```

```python
img_color = cv2.imread('data/verona_color.jpg')

fig_color, axes_color = plt.subplots(ncols=3, figsize=(3*6, 4), sharey=True)

# WARNING! OpenCV reads image default in "BGR"-mode, not "RGB"!

axes_color[0].hist(img_color[..., 2].flatten(), bins=256, color='red')
axes_color[1].hist(img_color[..., 1].flatten(), bins=256, color='green')
axes_color[2].hist(img_color[..., 0].flatten(), bins=256, color='blue');
```

We have written our own reusable function to plot image histograms. However, **it expects RGB-images, not BGR-images**, so make sure to do the conversion from BGR to RGB!

```python
from lib.plots import plot_img_hist

plot_img_hist(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
```

It also works with gray-scale images!

```python
plot_img_hist(img_gray)
```
