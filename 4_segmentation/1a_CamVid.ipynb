# CamVid dataset

For this example we will use the **CamVid** dataset. It consists of a train, validation and test subset. Each of these sets contains image-mask pairs of video frames taken on streets in the UK. The images have a height of 320 pixels and a width of 480 pixels. For more inforamtion about dataset visit http://mi.eng.cam.ac.uk/research/projects/VideoRec/CamVid/.

```python
from lib.camvid import download_camvid

DATA_DIR = './data/CamVid/'

download_camvid(DATA_DIR)
```

```python
from lib.camvid import CamVid

dataset = CamVid(DATA_DIR, 'train', classes=['car'])
```

```python
from lib.plot import visualize
from random import randint

image, mask = dataset[randint(0, len(dataset) - 1)] # get some sample
visualize(
    image=image, 
    cars_mask=mask.squeeze(),
)
```
