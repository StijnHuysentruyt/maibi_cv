# Depth map

## Load in the image and the corresponding depth map

A depth map of an image simply contains the depth of each pixel of the image. As such, it can also be represented as an image.

```python
import cv2

img = cv2.imread('data/livingroom_color.jpg')[..., ::-1]
depth = cv2.imread('data/livingroom_depth.png', cv2.IMREAD_UNCHANGED)

print(img.shape, depth.shape)
```

## Show the image and the depth map

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(ncols=2, figsize=(12, 4))

axes[0].imshow(img)
axes[1].imshow(depth)
```
